#**Step-3:Handling null values**
#delete the nan value
#deleting row or column
#impute the values(replace with mean,median,mode)

```python
#checking for null values
```

```python
df.isnull()
```

```python
df.isnull().sum()
```

#There are null values in "Age","Cabin" and "Embarked" columns

```python
#handling null values in "Age" column
#"Age" is a numerical column
df["Age"].fillna(df["Age"].mean(),inplace=True)
```

```python
df.isnull().sum()
```

```python
#Handling null values in "Cabin" column
df["Cabin"].fillna(df["Cabin"].mode()[0],inplace=True)
```

```python
df.isnull().sum()
```

```python
#Handling null values in "Embarked" column
df["Embarked"].fillna(df["Embarked"].mode()[0],inplace=True)
```

```python
df.isnull().sum()
```

#All the null valued columns were treated and made null value free

#Inference:There are outliers in "Age" column towards the upperlimit

```python
#Treating the outliers for "Age" attribute
q1=df.Age.quantile(0.25)
q3=df.Age.quantile(0.75)
```

```python
print(q1)
print(q3)
```

```python
IQR=q3-q1
IQR
```

```python
upper_limit=q3+1.5*IQR
upper_limit
```

```python
df.median()
```

```python
df['Age']=np.where(df['Age']>upper_limit,28,df['Age'])
```

```python
sns.boxplot(df.Age)
```

```python
sns.boxplot(df.SibSp)
```

```python
#Treating the outliers for "SibSp" attribute
from scipy import stats
```

```python
#x-mew/sigma value
SibSp_zscore=stats.zscore(df.SibSp)
SibSp_zscore
```

```python
#|x-mew/sigma| value
df_z=df[np.abs(SibSp_zscore)<=3]
```

```python
sns.boxplot(df_z.SibSp)
```
